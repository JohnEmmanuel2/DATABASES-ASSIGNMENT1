```python
import time
import os
import sys

filename = "example_output.txt"

def write_with_crash():
    data_chunks = [
        "Step 1: Writing header...\n",
        "Step 2: Writing body...\n",
        "Step 3: Writing more body...\n",
        "Step 4: Writing footer...\n"
    ]

    with open(filename, "w") as f:
        for i, chunk in enumerate(data_chunks):
            print(f"Writing chunk {i+1}")
            f.write(chunk)
            f.flush()          # force write to disk
            os.fsync(f.fileno())  # ensure OS flush

            time.sleep(0.5)

            if i == 1:
                print("Simulating crash now!")
                # Simulate a crash by killing the process
                os._exit(1)     # abrupt termination (no cleanup)

write_with_crash()
```
